## The Object Metaphor

```python
from datetime import date
```

```python
tues = date(2026, 1, 14)
print(date(2026, 11, 18) - tues)
```

```python
tues.year
```

```python
tues.strftime('%A, %B %d')
# A是星期名
```

## Sequence Objects

```python
chinese = ['coin', 'string', 'myriad'] 
suits = chinese 
suits.pop()
suits.remove("string")
```

```python
suits.append('cup')
suits.extend(['sword', 'club'])
suits[2] = 'spade'
```
