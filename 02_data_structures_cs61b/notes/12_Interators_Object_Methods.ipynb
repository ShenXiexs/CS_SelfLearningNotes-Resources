### Set

Set 是一组唯一元素的集合，也就是说，每个元素只能出现一次。

和 List 不同，Set 没有顺序概念：

你不能通过下标访问集合中的元素
你也不能控制元素插入到集合中的哪个位置
Java 中的 Set

Java 提供了 Set 接口及其实现类，例如 HashSet。

如果你不想写完整类名，也记得先导入它们：

```java
import java.util.Set;
import java.util.HashSet;

Set<String> S = new HashSet<>();
S.add("Hello");
S.add("Shenzhen");
System.out.println(S);
```

```java
import java.util.Set;
import java.util.HashSet;

Set<String> s = new HashSet<>();
s.add("Tokyo");
s.add("Lagos");

System.out.println(s.contains("Tokyo"));
System.out.println(s);
```

我们的目标是实现自己的集合 ArraySet，并支持以下方法：

add(value)：如果该值还不存在，就把它加入集合
contains(value)：检查 ArraySet 是否包含这个键
size()：返回集合中元素的数量

```java
class ArraySet {
    private String[] items;
    private int size;

    public ArraySet() {
        items = new String[100];
        size = 0;
    }

    public boolean contains(String x) {
        for (int i = 0; i < size; i++) {
            if (items[i].equals(x)) {
                return true;
            }
        }
        return false;
    }

    public void add(String x) {
        if (x == null) {
            return;
        }

        if (!contains(x)) {
            items[size] = x;
            size += 1;
        }
    }

    public int size() {
        return size;
    }

    public void printSet() {
        System.out.println("The set contains:");
        for (int i = 0; i < size; i++) {
            System.out.println(items[i] + " ");
        }
        System.out.println(".");
    }
}
```

```java
ArraySet s = new ArraySet();
s.add("Tokyo");
s.add("Lagos");
s.add("Tokyo");

System.out.println(s.contains("Tokyo"));
System.out.println(s.contains("Beijing"));
System.out.println(s.size());
s.printSet();
```

### 12.2 异常（Exceptions）

在这一节中，我们将学习如何通过**抛出异常（throw exceptions）**来有效处理代码中可能出现的错误。

我们在上一节实现的 ArraySet 存在一个小问题：
当我们向 ArraySet 中添加 null 时，会得到一个 NullPointerException。

问题出在 contains 方法中，我们执行了 items[i].equals(x)。
如果 items[i] 的值是 null，那么实际上是在调用：

null.equals(x)

这就会导致 NullPointerException。

异常会使程序的正常执行流程中断。
实际上，我们也可以选择主动抛出自己的异常。

在 Python 中，你可能见过使用 raise 关键字来实现这一点。
在 Java 中，异常是对象，我们使用如下格式来抛出异常：

throw new ExceptionObject(parameter1, ...)

现在，我们来处理一个场景：
当用户尝试向 ArraySet 中添加 null 时，我们主动抛出异常。

我们将抛出一个 IllegalArgumentException，它接受一个参数（字符串类型的错误信息）。

```java
/* 将指定的值关联到该映射中。
   如果 key 为 null，则抛出 IllegalArgumentException。 */
public void add(T x) {
    if (x == null) {
        throw new IllegalArgumentException("can't add null");
    }
    if (contains(x)) {
        return;
    }
    items[size] = x;
    size += 1;
}
```

### 12.3 迭代（Iteration）

我们可以在 Java 的 HashSet 上使用一种简洁的增强 for 循环（foreach）：

```java
import java.util.Set;
import java.util.HashSet;

Set<String> S = new HashSet<>();
S.add("Hello");
S.add("World");
for (String s : S) {
    System.out.println(s);
}
```

核心概念：Iterator（迭代器）

关键在于一个对象：迭代器（Iterator）

例如，在 List 中，我们会定义：

```java
public Iterator<E> iterator();

import java.util.*;

List<Integer> friends = new ArrayList<>();
friends.add(1);
friends.add(2);
friends.add(3);

Iterator<Integer> seer = friends.iterator();

while (seer.hasNext()) {
    System.out.println(seer.next());
}
```

Iterator 的三个核心操作

获取迭代器

Iterator<Integer> seer = friends.iterator();

判断是否还有元素

seer.hasNext()

获取下一个元素（同时向前移动）

seer.next()

#### 为 ArraySet 实现 Iterator

我们在 ArraySet 内部写一个嵌套类：

```java
import java.util.Iterator;

class ArraySet<T> implements Iterable<T> {
    private T[] items;
    private int size;

    public ArraySet() {
        items = (T[]) new Object[100];
        size = 0;
    }

    public boolean contains(T x) {
        for (int i = 0; i < size; i++) {
            if (items[i].equals(x)) {
                return true;
            }
        }
        return false;
    }

    public void add(T x) {
        if (x == null) {
            throw new IllegalArgumentException("can't add null");
        }
        if (contains(x)) {
            return;
        }
        items[size] = x;
        size++;
    }

    public int size() {
        return size;
    }

    public Iterator<T> iterator() {
        return new ArraySetIterator();
    }

    private class ArraySetIterator implements Iterator<T> {
        private int wizPos;

        public ArraySetIterator() {
            wizPos = 0;
        }

        public boolean hasNext() {
            return wizPos < size;
        }

        public T next() {
            T returnItem = items[wizPos];
            wizPos++;
            return returnItem;
        }
    }
}
```

```java
ArraySet<Integer> aset = new ArraySet<>();
aset.add(5);
aset.add(23);
aset.add(42);

for (int i : aset) {
    System.out.println(i);
}
```
