```java
public class Dog {
    public int weightInPounds;
    public void makeNoise() {
        if (weightInPounds < 10) {
            System.out.println("Yip yip yip!");
        } else if (weightInPounds < 30) {
            System.out.println("Bark bark bark!");
        } else {
            System.out.println("Woof woof woof!");
        }
    }
}

public class DogLauncher {
    public static void main(String[] args) {
        Dog tiny = new Dog();
        tiny.weightInPounds = 5;
        tiny.makeNoise(); // This will print "Yip yip yip!" to the console.

        Dog medium = new Dog();
        medium.weightInPounds = 20;
        medium.makeNoise(); // This will print "Bark bark bark!" to the console.

        Dog large = new Dog();
        large.weightInPounds = 50;
        large.makeNoise(); // This will print "Woof woof woof!" to the console.
    }
}

DogLauncher.main(null); // This will execute the main method and print the noises for tiny, medium, and large dogs to the console.

public class DogLauncher {
    public static void main(String[] args) {
        Dog d;
        d = new Dog();
        d.weightInPounds = 20;
        d.makeNoise();
    }
}
DogLauncher.main(null); // This will print "Bark bark bark!" to the console.
```

```java
public class Dog {
    public int weightInPounds;

    public Dog(int w) {
        weightInPounds = w;
    }  

    public void makeNoise() {
        if (weightInPounds < 10) {
            System.out.println("Yip yip yip!");
        } else if (weightInPounds < 30) {
            System.out.println("Bark bark bark!");
        } else {
            System.out.println("Woof woof woof!");
        }
    }
    public static Dog maxDog(Dog d1, Dog d2) {
    if (d1.weightInPounds > d2.weightInPounds) {
        return d1;
    }
    return d2;
}
}


Dog d1 = new Dog(15);
Dog d2 = new Dog(100);
Dog.maxDog(d1, d2);
/*REPL.$JShell$12K$Dog@388001cd*/
System.out.println(Dog.maxDog(d1, d2).weightInPounds); // This will print 100 to the console.
```
