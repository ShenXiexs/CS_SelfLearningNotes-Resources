## 测试与选择排序（Testing and Selection Sort）

对于一个中级到高级程序员来说，最重要的技能之一，就是判断自己的代码是否正确的能力。本章将讨论如何编写测试来评估代码的正确性。在这个过程中，我们还会介绍一种排序算法，叫做选择排序（Selection Sort）。

### 一种新的思路（A New Way）

当你写程序时，代码可能会有错误。在课堂环境中，你通常会通过用户交互、代码分析和自动评测器（autograder）测试这几种方式的结合，来建立对代码正确性的信心。在很多情况下，最后这一项尤其重要，因为它往往决定了你能不能拿到分数。

当然，自动评测器并不是什么魔法。它本质上也是老师写出来的一段代码，和你自己写的代码并没有根本区别。现实世界里，这些测试通常都是程序员自己写的，而不是由某个善良的、像 Josh Hug 一样的第三方来替你完成。

本章中，我们将探索如何自己编写测试。我们的目标是创建一个名为 Sort 的类，它提供一个方法 sort(String[] x)，这个方法会**原地（destructively）**对数组 x 中的字符串进行排序。

作为一种全新的思考方式，我们会先写 testSort()，在把测试写完之后，再去编写真正的排序代码。

### 临时式测试（Ad Hoc Testing）

给 Sort.sort 写测试其实很直接，只是会比较繁琐。我们只需要：

创建输入；
调用 sort；
检查输出是否正确。

如果输出不正确，我们就打印第一个不匹配的位置，并结束测试。比如，我们可以这样写一个测试类：

```java
public class TestSort {
    /** 测试Sort 类的 sort方法*/
    public static void testSort() {
        String[] input = {"i", "have", "an", "apple"};
        String[] expected = {"an", "apple", "have", "i"};
        Sort.sort(input);

        for (int i = 0; i < input.length; i += 1) {
            if (!input[i].equals(expected[i])) {
                System.out.println(
                    "Mismatch in position " + i + ": expected " + expected[i] + ", but got " + input[i]
                );
                break;
            }
        }
    }

    public static void main(String[] args) {
        testSort();
    }
}
```

一个重要说明：你可能会问：“为什么要遍历整个数组？为什么不用 == 直接比较两个数组是否相等？”

原因是：当我们比较两个对象是否相等时，不能简单地使用 ==。
== 比较的是内存盒子里的“字面位模式”，也就是说：

input == expected

**测试的是 input 和 expected 这两个引用是否指向同一个地址，而不是它们数组里的值是否相同。**

因此，在 testSort 中，我们使用了循环，并打印第一个不匹配的位置。你当然也可以使用 Java 内置的方法 java.util.Arrays.equals 来代替循环。

虽然上面的这个单独测试写起来还不算太麻烦，但如果要写一整套这样的临时测试，就会变得非常繁琐，因为你得反复写很多循环和打印语句。下一节我们会看到，org.junit 库是如何帮我们省掉很多麻烦的。

```java
import static com.google.common.truth.Truth.assertThat;

public class TestSort {
    /** 测试 Sort 类的 sort 方法。 */
    public static void testSort() {
        String[] input = {"cows", "dwell", "above", "clouds"};
        String[] expected = {"above", "clouds", "cows", "dwell"};
        Sort.sort(input);

        assertThat(input).isEqualTo(expected);
    }

    public static void main(String[] args) {
        testSort();
    }
}
```
