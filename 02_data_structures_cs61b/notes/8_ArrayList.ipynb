# 8. ArrayList

https://cs61b-2.gitbook.io/cs61b-textbook/8.-arraylist<br>
在这一节中，我们将构建一个新的类 AList，用于存储任意长度的数据列表，类似于我们之前实现的 DLList。
不同的是，AList 使用数组存储数据，而不是链表。

## Linked List Performance Puzzle（链表性能问题）

假设我们在 DLList 中写一个新方法：

int get(int i)

为什么对于很长的链表，get 会比 getLast 慢？
在什么情况下会特别慢？

Puzzle 解释

原因是：

在双向链表中，我们只有：

指向第一个节点的引用
指向最后一个节点的引用

如果我们想访问第 i 个元素，就必须：

从前往后走 i 步
或者从后往前走 (size - i) 步

例如：
如果链表长度 = 10000，要访问第 417 个元素，就必须走 417 步。

最坏情况：元素在中间，需要走大约 N/2 步。
因此：

方法	时间复杂度
get(i)	O(N)
getLast()	O(1)

也就是说：

get 是线性时间
getLast 是常数时间

### Our First Attempt: Naive Array Based List

数组访问第 i 个元素是 常数时间，因为可以直接通过下标访问：

items[i]

因此：
数组实现的列表在 get 操作上比链表快很多。

### AList 的不变量（非常重要）

AList 实现通常保持以下规则：

下一个插入位置 = size
列表元素数量 = size
最后一个元素位置 = size - 1

这三个规则叫做 invariants（不变量）

非常重要。

### removeLast

我们需要实现 removeLast 操作。

关键思想：

列表的任何变化，都必须反映在数组或 size 的变化上。

因为：

列表是抽象概念
数组和 size 是具体实现

所以 removeLast 应该：

size--
把删除位置设为 null（防止内存泄漏）

### Naive Resizing Arrays（朴素扩容）

如果数组满了怎么办？

Java 数组大小不能改变，所以要：

创建新数组 → 复制旧数据 → 指向新数组

示例代码：

int[] a = new int[size + 1];
System.arraycopy(items, 0, a, 0, size);
a[size] = 11;
items = a;
size = size + 1;

这个过程叫：

Resizing（扩容）

### Naive Resizing 的问题

如果每次 addLast 都新建数组：

时间复杂度会变成：

O(N²)

因为每次都要复制整个数组。

插入 100000 个元素会非常慢。

### Geometric Resizing（几何扩容）

正确做法不是每次 +1，而是 乘以一个倍数。

例如：

满了 → 扩容为 2 倍
这样扩容次数很少
平均 addLast 时间 = O(1)

错误方式：

resize(size + 1);

正确方式：

resize(size * 2);

### Memory Performance（内存优化）

如果数组很大但元素很少，会浪费内存。

解决方法：
当使用率过低时缩小数组。

定义：

usage ratio = size / items.length

如果：

usage ratio < 0.25

就把数组缩小一半。

### Generic ALists（泛型 AList）

我们可以让 AList 存储任意类型，而不是 int。

Java 泛型数组不能这样写：

T[] items = new T[8];   // 错误

必须这样写：

T[] items = (T[]) new Object[8];

删除元素时必须：

items[index] = null;

否则 Java 垃圾回收不会释放对象（loitering）。
