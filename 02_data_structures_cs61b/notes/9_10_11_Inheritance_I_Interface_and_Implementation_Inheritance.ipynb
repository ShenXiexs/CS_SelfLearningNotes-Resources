```java
interface List61B<Item> {
    void addLast(Item x);
    Item get(int i);
    int size();
}

class SLList<Item> implements List61B<Item> {
    private class Node {
        Item item;
        Node next;
        Node(Item i, Node n) {
            item = i;
            next = n;
        }
    }

    private Node first;
    private int size;

    public SLList() {
        first = null;
        size = 0;
    }

    public void addLast(Item x) {
        if (first == null) {
            first = new Node(x, null);
        } else {
            Node p = first;
            while (p.next != null) {
                p = p.next;
            }
            p.next = new Node(x, null);
        }
        size++;
    }

    public Item get(int i) {
        Node p = first;
        for (int j = 0; j < i; j++) {
            p = p.next;
        }
        return p.item;
    }

    public int size() {
        return size;
    }
}
```

```java
interface List61B<Item> {
    void addFirst(Item x);
    void addLast(Item x);
    Item getFirst();
    Item getLast();
    Item removeLast();
    Item get(int i);
    int size();

    default void print() {
        for (int i = 0; i < size(); i++) {
            System.out.print(get(i) + " ");
        }
        System.out.println();
    }
}
```

```java
class AList<Item> implements List61B<Item> {
    private Item[] items;
    private int size;

    public AList() {
        items = (Item[]) new Object[8];
        size = 0;
    }

    private void resize(int capacity) {
        Item[] newItems = (Item[]) new Object[capacity];
        System.arraycopy(items, 0, newItems, 0, size);
        items = newItems;
    }

    public void addFirst(Item x) {
        insert(x, 0);
    }

    public void addLast(Item x) {
        if (size == items.length) {
            resize(items.length * 2);
        }
        items[size] = x;
        size++;
    }

    public void insert(Item x, int position) {
        if (size == items.length) {
            resize(items.length * 2);
        }
        for (int i = size; i > position; i--) {
            items[i] = items[i - 1];
        }
        items[position] = x;
        size++;
    }

    public Item getFirst() {
        return items[0];
    }

    public Item getLast() {
        return items[size - 1];
    }

    public Item removeLast() {
        Item x = items[size - 1];
        items[size - 1] = null;
        size--;
        return x;
    }

    public Item get(int i) {
        return items[i];
    }

    public int size() {
        return size;
    }
}
```

```java
class SLList<Item> implements List61B<Item> {
    private class Node {
        Item item;
        Node next;
        Node(Item i, Node n) {
            item = i;
            next = n;
        }
    }

    private Node first;
    private int size;

    public void addFirst(Item x) {
        first = new Node(x, first);
        size++;
    }

    public void addLast(Item x) {
        if (first == null) {
            addFirst(x);
            return;
        }
        Node p = first;
        while (p.next != null) {
            p = p.next;
        }
        p.next = new Node(x, null);
        size++;
    }

    public Item getFirst() {
        return first.item;
    }

    public Item getLast() {
        Node p = first;
        while (p.next != null) {
            p = p.next;
        }
        return p.item;
    }

    public Item removeLast() {
        Node p = first;
        if (p.next == null) {
            Item x = p.item;
            first = null;
            size--;
            return x;
        }
        while (p.next.next != null) {
            p = p.next;
        }
        Item x = p.next.item;
        p.next = null;
        size--;
        return x;
    }

    public Item get(int i) {
        Node p = first;
        for (int j = 0; j < i; j++) {
            p = p.next;
        }
        return p.item;
    }

    public int size() {
        return size;
    }
}
```

### 11.2 子类型多态 vs 显式高阶函数

我们已经看到，继承可以让我们在父类中复用已有代码，同时通过重写父类方法或在子类中编写新方法来实现一些小的修改。继承还使我们能够利用多态来设计通用的数据结构和方法。

多态（Polymorphism）从本质上讲，意味着“多种形态”。
在 Java 中，多态指的是对象可以具有多种形式或类型。

在面向对象编程中，一个对象可以被看作是：

它自身类的实例
它父类的实例
它祖先类的实例
以及更高层父类的实例（依次类推）

考虑一个静态类型为 Deque 的变量 deque。
当调用 deque.addFirst() 时，实际执行哪个方法是在运行时决定的，取决于 deque 的运行时类型（即动态类型）。

正如上一章所讲，Java 是通过**动态方法选择（dynamic method selection）**来决定调用哪个方法的。

假设我们想写一个 Python 程序，用来打印两个对象中“较大者”的字符串表示。有两种实现方式：

```java
def print_larger(x, y, compare, stringify):
    if compare(x, y):
        return stringify(x)
    return stringify(y)

def print_larger(x, y):
    if x.largerThan(y):
        return x.str()
    return y.str()
```

使用显式高阶函数的方法，你提供了一种通用的方式来比较并打印两个对象中较大的那个。

而在子类型多态的方法中，由对象自身来决定如何比较和如何转换为字符串。

也就是说，调用哪个 largerThan 方法，取决于 x 和 y 的实际类型。

### Comparable vs Comparator（核心区别）
Comparable：

“我自己会比较自己”

比较逻辑写在对象内部

定义的是自然顺序（natural ordering）

Comparator：

“我作为第三方来比较两个对象”

比较逻辑在对象外部

可以定义多种排序方式

### 11.5 本章总结

回顾：类型规则

编译器允许“内存盒子”存储任意子类型。

编译器根据静态类型来允许方法调用。

被重写的非静态方法在运行时根据动态类型选择。

对于重载方法，方法是在编译期选择的。

子类型多态  
考虑一个静态类型为 Deque 的变量。调用 deque.method() 的行为取决于其动态类型。因此，我们可以有许多实现 Deque 接口的子类，它们都可以调用 deque.method()。

子类型多态示例  
假设我们想编写一个函数 max()，用于返回任意数组中的最大值，而不考虑数组中元素的类型。如果我们写一个方法 max(Object[] items)，并使用 > 运算符来比较数组中的每个元素，这将无法工作。为什么？

这是因为这种做法假设所有对象都可以进行比较，但实际上有些对象是不能比较的。另一种方法是为 Dog 类编写一个 max() 函数，但这样一来，对于每一个需要比较的类，我们都必须写一个 max() 函数。请记住，我们的目标是编写一个“唯一的 max 方法”，能够适用于所有可比较的对象。

解决方案：OurComparable 接口  
解决方法是创建一个包含 compareTo(Object) 方法的接口，我们称之为 OurComparable。现在，我们的 max() 方法可以接受一个 OurComparable[] 参数。由于我们保证任何实现该接口的对象都具备接口中的所有方法，因此我们可以确保始终能够调用 compareTo 方法，并且该方法能够正确返回对象之间的顺序关系。

现在，我们就可以定义一个“唯一的 max 方法”。当然，任何需要被比较的对象都必须实现 compareTo 方法。不过，我们不再需要在每个类中重复实现 max 的逻辑，而只需要实现两个对象之间如何比较大小的逻辑即可。

更好的方案：Java 内置的 Comparable  
Java 提供了内置的 Comparable 接口，并使用泛型来避免一些奇怪的类型转换问题。此外，Comparable 已经适用于 Integer、Character 和 String 等类型，而且这些类型已经为你实现了 max、min 等方法。因此，你不需要重复已经完成的工作。

Comparators  
“自然顺序”这一术语指的是由 Comparable 的 compareTo 方法所定义的排序方式。然而，如果我们希望按照除 size 之外的其他方式对 Dog 对象进行排序，该怎么办？

此时，我们可以传入一个 Comparator<T> 接口，该接口要求实现 compare() 方法。然后，我们可以按照需要自由实现 compare() 方法，从而实现我们想要的排序方式。
