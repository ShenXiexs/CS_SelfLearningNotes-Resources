```java
interface List61B<Item> {
    void addLast(Item x);
    Item get(int i);
    int size();
}

class SLList<Item> implements List61B<Item> {
    private class Node {
        Item item;
        Node next;
        Node(Item i, Node n) {
            item = i;
            next = n;
        }
    }

    private Node first;
    private int size;

    public SLList() {
        first = null;
        size = 0;
    }

    public void addLast(Item x) {
        if (first == null) {
            first = new Node(x, null);
        } else {
            Node p = first;
            while (p.next != null) {
                p = p.next;
            }
            p.next = new Node(x, null);
        }
        size++;
    }

    public Item get(int i) {
        Node p = first;
        for (int j = 0; j < i; j++) {
            p = p.next;
        }
        return p.item;
    }

    public int size() {
        return size;
    }
}
```

## Method Overloading

### Hypernyms, Hyponyms（上位词 & 下位词）

这是一个非常重要的思想。

例子：

单词	 关系
Dog	    上位词
Poodle	下位词
Husky	下位词

关系叫：

is-a 关系
Poodle is-a Dog
Dog is-a Animal
在 Java 里

同样的关系：

SLList is-a List
AList  is-a List

所以我们可以创建一个更通用的类型：List61B

然后：

SLList implements List61B
AList  implements List61B
Interface（接口）

接口就是一个：

行为合同（contract）

它只规定：

必须有哪些方法
不规定怎么实现

```java
interface List61B<Item> {
    void addFirst(Item x);
    void addLast(Item x);
    Item getFirst();
    Item getLast();
    Item removeLast();
    Item get(int i);
    int size();

    default void print() {
        for (int i = 0; i < size(); i++) {
            System.out.print(get(i) + " ");
        }
        System.out.println();
    }
}
```

```java
class AList<Item> implements List61B<Item> {
    private Item[] items;
    private int size;

    public AList() {
        items = (Item[]) new Object[8];
        size = 0;
    }

    private void resize(int capacity) {
        Item[] newItems = (Item[]) new Object[capacity];
        System.arraycopy(items, 0, newItems, 0, size);
        items = newItems;
    }

    public void addFirst(Item x) {
        insert(x, 0);
    }

    public void addLast(Item x) {
        if (size == items.length) {
            resize(items.length * 2);
        }
        items[size] = x;
        size++;
    }

    public void insert(Item x, int position) {
        if (size == items.length) {
            resize(items.length * 2);
        }
        for (int i = size; i > position; i--) {
            items[i] = items[i - 1];
        }
        items[position] = x;
        size++;
    }

    public Item getFirst() {
        return items[0];
    }

    public Item getLast() {
        return items[size - 1];
    }

    public Item removeLast() {
        Item x = items[size - 1];
        items[size - 1] = null;
        size--;
        return x;
    }

    public Item get(int i) {
        return items[i];
    }

    public int size() {
        return size;
    }
}
```

```java
class SLList<Item> implements List61B<Item> {
    private class Node {
        Item item;
        Node next;
        Node(Item i, Node n) {
            item = i;
            next = n;
        }
    }

    private Node first;
    private int size;

    public void addFirst(Item x) {
        first = new Node(x, first);
        size++;
    }

    public void addLast(Item x) {
        if (first == null) {
            addFirst(x);
            return;
        }
        Node p = first;
        while (p.next != null) {
            p = p.next;
        }
        p.next = new Node(x, null);
        size++;
    }

    public Item getFirst() {
        return first.item;
    }

    public Item getLast() {
        Node p = first;
        while (p.next != null) {
            p = p.next;
        }
        return p.item;
    }

    public Item removeLast() {
        Node p = first;
        if (p.next == null) {
            Item x = p.item;
            first = null;
            size--;
            return x;
        }
        while (p.next.next != null) {
            p = p.next;
        }
        Item x = p.next.item;
        p.next = null;
        size--;
        return x;
    }

    public Item get(int i) {
        Node p = first;
        for (int j = 0; j < i; j++) {
            p = p.next;
        }
        return p.item;
    }

    public int size() {
        return size;
    }
}
```

## WordUtils

```java
class WordUtils {
    public static String longest(List61B<String> list) {
        int maxDex = 0;
        for (int i = 0; i < list.size(); i++) {
            if (list.get(i).length() > list.get(maxDex).length()) {
                maxDex = i;
            }
        }
        return list.get(maxDex);
    }
}
```

```java
public class Main {
    public static void main(String[] args) {
        List61B<String> list1 = new SLList<>();
        list1.addLast("cat");
        list1.addLast("elephant");
        list1.addLast("dog");

        List61B<String> list2 = new AList<>();
        list2.addLast("apple");
        list2.addLast("banana");
        list2.addLast("watermelon");

        System.out.println(WordUtils.longest(list1));
        System.out.println(WordUtils.longest(list2));

        list1.print();
        list2.print();
    }
}
```

```java
public class RotatingSLList<Item> extends SLList<Item> {
    public void rotate() {
        if (size() <= 1) {
            return;
        }
        Item firstItem = getFirst();
        removeFirst();
        addLast(firstItem);
    }
}
```
