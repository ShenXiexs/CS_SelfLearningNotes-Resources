```java
public class ConditionalsWithBlocks {
    public static void main(String[] args) {
        int x = 5;
        if (x<10){
            System.out.println("I shall increase x by 10");
            x = x + 10;            
        }

        if (x<10){
            System.out.println("I shall increase x by 10");
            x = x + 10;
        }
        System.out.println(x);
    }
}
ConditionalsWithBlocks.main(null);
```

```java
public class PrintAbsoluteValue {
    public static void main(String[] args) {
        int x = 5;

        if (x < 0)
            System.out.println("I should negate X");
            x = -x;

        System.out.println(x);
    }
}
PrintAbsoluteValue.main(null);
/*
在 Java 中，花括号非常重要！与 Python 不同，
Java 语句是用花括号而不是缩进来分组的
为了演示这样做会造成哪些严重问题，请尝试运行以下程序（或使用此链接），
该程序应该打印出 的绝对值 x。然后尝试将x的值更改x为一个正数。运行程序并确保你理解为什么会出错。
!!!如果 没有花括号 {}，if 只控制 紧接着的一条语句!!!
*/

public class PrintAbsoluteValue2 {
    public static void main(String[] args) {
        int x = 5;

        if (x < 0){
            System.out.println("I should negate X");
            x = -x;            
        }


        System.out.println(x);
    }
}
PrintAbsoluteValue2.main(null);
```

```java
String a = "Achilles";
String t = "Tortoise";
double aPos = 0;
double tPos = 100;
double aSpeed = 20;
double tSpeed = 10;
double totalTime = 0;
while (aPos < tPos) { 
    System.out.println("At time: " + totalTime);
    System.out.println("    " + a + " is at position " + aPos);
    System.out.println("    " + t + " is at position " + tPos);

    double timeToReach = (tPos - aPos) / aSpeed;
    totalTime = totalTime + timeToReach;
    aPos = aPos + timeToReach * aSpeed;
    tPos = tPos + timeToReach * tSpeed;
}
```
