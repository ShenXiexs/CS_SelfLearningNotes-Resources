```python
def absolute_value(x):
    if x >= 0:
        x = x
    else:
        x = -x
    return(x)

absolute_value(-4)
absolute_value(5)
absolute_value(0)
```

```python
def fib(n):
    if not isinstance(n, int):
        print("Wrong n. n should be int")
        return
    elif n==1:
        n=0
    else:
        pred, curr = 0,1
        k = 2
        while(k<n):
            pred, curr = curr, pred+curr
            k+=1
        return(curr)

fib(8)
```

```python
def  fact_iter ( n ): 
    total, k = 1 , 1 
    while k <= n : 
        total , k = total * k , k + 1 
    return total
fact_iter(6)
```

```python
def  cascade ( n ): 
    if n < 10 : 
        print ( n ) 
    else : 
        print ( n ) 
        cascade ( n // 10 ) 
        print ( n ) 
cascade(10)
```

```python
def  count_partitions ( n ,  m ): 
    """计算用最多 m 个部分将 n 分割成若干个部分的方法数。"""
    if n==0:
        return 1
    elif n<0:
        return 0
    elif m==0:
        return 0
    else:
        return count_partitions ( n-m ,  m ) +  + count_partitions ( n , m - 1 )  
count_partitions ( 10 ,  2 )
```
